```python
import numpy as np
import matplotlib.pyplot as plt
import py4DSTEM
```

```python
fp = "/home/ben/Data/20190128_ClassificationData/EASY - Frances Au grains from Dow.dm4"
dc = py4DSTEM.file.readwrite.read(fp)
```

```python
dc.swap_RQ()
```

```python
dc.data4D.shape
```

```python
ave_im = np.average(dc.data4D,axis=(2,3))
```

```python
rx=(20,40,60)
ry=(40,100,150)

fig,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,figsize=(12,12))
ax11.matshow(ave_im)
ax11.scatter(ry,rx, color=('r','y','b'))
ax12.matshow(dc.data4D[rx[0],ry[0],:,:]**0.1)
ax21.matshow(dc.data4D[rx[1],ry[1],:,:]**0.1)
ax22.matshow(dc.data4D[rx[2],ry[2],:,:]**0.1)
plt.show()
```
